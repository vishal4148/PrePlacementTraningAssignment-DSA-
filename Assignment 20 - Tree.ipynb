#### Question-1

Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :

   1

 /   \

2      3

/ \    / \

4   5  6   7

```python
# Global variables
max_sum = float('-inf')
max_subtree = None

def findSubtreeSum(root):
    global max_sum, max_subtree
    
    if root is None:
        return 0
    
    left_sum = findSubtreeSum(root.left)
    right_sum = findSubtreeSum(root.right)
    
    subtree_sum = root.val + left_sum + right_sum
    
    if subtree_sum > max_sum:
        max_sum = subtree_sum
        max_subtree = root
    
    return subtree_sum

# Test the algorithm
# Define the binary tree
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build the tree from the example

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)


# Find the subtree with the maximum sum
findSubtreeSum(root)

# Print the maximum sum and the corresponding subtree root value
print("Maximum Sum:", max_sum)
print("Subtree Root:", max_subtree.val)
```

#### Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

        7

     /    \

    4      12

  /  \      /

  3   6     8

 /    /     \

1    5      10

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBST(level_order):
    if not level_order:
        return None

    root = TreeNode(level_order[0])
    queue = [root]
    index = 1

    while queue and index < len(level_order):
        current = queue.pop(0)

        left_val = level_order[index] if index < len(level_order) else None
        right_val = level_order[index + 1] if index + 1 < len(level_order) else None

        if left_val is not None:
            current.left = TreeNode(left_val)
            queue.append(current.left)

        if right_val is not None:
            current.right = TreeNode(right_val)
            queue.append(current.right)

        index += 2

    return root

def inorderTraversal(root):
    if root is None:
        return []

    result = []
    stack = []

    while root or stack:
        while root:
            stack.append(root)
            root = root.left

        root = stack.pop()
        result.append(root.val)
        root = root.right

    return result

# Test the algorithm
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(level_order)

# Perform inorder traversal to verify the constructed BST
inorder = inorderTraversal(root)
print("Inorder Traversal:", inorder)
```

#### Question-3

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

```python
from collections import deque

def isLevelOrderBST(arr):
    if not arr:
        return "Yes"

    n = len(arr)
    nodes = deque()
    min_max = deque()
    root = arr[0]
    nodes.append(root)
    min_max.append((float('-inf'), float('inf')))

    i = 1
    while i < n:
        node = nodes.popleft()
        min_val, max_val = min_max.popleft()

        if arr[i] < min_val or arr[i] > max_val:
            return "No"

        if arr[i] < node:
            nodes.append(arr[i])
            min_max.append((min_val, node - 1))

        if arr[i] > node:
            nodes.append(arr[i])
            min_max.append((node + 1, max_val))

        i += 1

    return "Yes"

# Test the algorithm
arr = [7, 4, 12, 3, 6, 5, 1, 5, 10]
result = isLevelOrderBST(arr)
print("Result:", result)
```
